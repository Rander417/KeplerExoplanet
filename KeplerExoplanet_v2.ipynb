# Data Load

```python
# Load dataset
file_path = "./Resources/cumulative.csv"
keplerRAW_df = pd.read_csv(file_path)
keplerRAW_df
```

```python
# change the index to the "kepoi_name" and drop the title
keplerProcessed_df = keplerRAW_df.set_index('kepoi_name')
keplerProcessed_df.rename_axis(None, inplace=True)

keplerProcessed_df 
```

```python
# Both of these columns are empty for this dataset
keplerProcessed_df.drop(["koi_teq_err1","koi_teq_err2"], axis =1, inplace=True)
keplerProcessed_df
```

### <span style="color:red"> Dropping all NaNs here to get model running. Removes a large amount of rows. Need to Address </span>

```python
# Drop rows with Nan
keplerProcessed_df = keplerProcessed_df.dropna()
keplerProcessed_df
```

```python
# Store relationships between the id, name, and dispositions 
keplerID_Names_df = keplerProcessed_df[["kepid","kepler_name","koi_disposition","koi_pdisposition"]]
keplerID_Names_df
```

```python
# Drop columns that do not add value to the model

# koi_teq_err1 & koi_teq_err2 are completely NaN for this dataset
keplerProcessed_df.drop(["rowid","kepid","kepler_name","koi_disposition"], axis =1, inplace=True)
keplerProcessed_df
```

```python
# Convert koi_tce_delivname to numeric dummies
keplerProcessed_df = pd.get_dummies(keplerProcessed_df, columns=["koi_tce_delivname"])
keplerProcessed_df
```

### <span style="color:red">NOTE: I'm using the koi_pdisposition for simplification since it has (2) values. Lets look at what it takes to use koi_disposition instead (has 4 values)</span>

```python
# Convert koi_pdisposition to numeric, this will be the "target"
le = LabelEncoder()

keplerProcessed_df['koi_pdisposition'] = le.fit_transform(keplerProcessed_df['koi_pdisposition'])
keplerProcessed_df
```

### Separate the Features (X) from the Target (y)

```python
y = keplerProcessed_df["koi_pdisposition"]
X = keplerProcessed_df.drop(["koi_pdisposition"], axis =1)
```

### Split our data into training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

# Supervised ML Logistic Regression Model

```python
# Create the model
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

```python
# Train model
classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

# BASIC NEURAL NET w/ SIGMOID

### Preprocess data for neural network (scaling)

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Build, Train and evaluate model

```python
# Define the basic neural network model
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu", input_dim=42))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)

# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
